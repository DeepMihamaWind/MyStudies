# Chapter 5(p.143) ネイピア数を確認する

$\lim\limits_{h \to 0} (1 + h) ^\frac{1}{h} = e$

```
import numpy as np
```

```
np.set_printoptions(precision = 10)
```

```
x = np.logspace(0, 11, 12, base=0.1, dtype='float64')
x
```

```
y = np.power(1+x, 1/x)
y
```

```
lines = len(x)
for line in range(lines):
    print('y = %.11f h = %.11f' %(y[line], x[line])) # p.143 図5-8では11行しか表示されていなかったので改変
```

```
e = np.e
```

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(y, linestyle =':' ,color='b',label = '$(1 + h)^\\frac{1}{h}$')
plt.plot([0,11],[e,e],color='r',label = '$e$')
plt.xlabel('$(\\frac{1}{10})^n$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```
